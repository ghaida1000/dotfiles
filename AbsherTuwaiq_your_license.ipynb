```python
import os
from PIL import Image
import pprint

def get_image_size(image_path):
    with Image.open(image_path) as img:
        return img.width, img.height

def convert_to_yolo_format(json_file, output_dir, images_dir):
    with open(json_file, 'r') as f:
        data = json.load(f)

    # Determine the structure of the JSON
    if isinstance(data, dict):
        if 'annotations' in data:
            annotations = data['annotations']
        elif 'data' in data:
            annotations = data['data']
        else:
            raise KeyError("JSON structure is not recognized. Expected keys like 'annotations' or 'data'.")
    elif isinstance(data, list):
        annotations = data
    else:
        raise TypeError("JSON root is neither a list nor a dictionary.")

    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Optional: Define class mapping if necessary
    # class_mapping = {"road": 0, "sidewalk": 1, "vehicle": 2}  # Example

    for annotation in annotations:
        if not isinstance(annotation, dict):
            print("Skipping annotation because it's not a dictionary:", annotation)
            continue

        try:
            image_id = annotation['image_id']
            objects = annotation['objects']
            image_filename = f"{image_id}.jpg"  # Adjust extension if different
            image_path = os.path.join(images_dir, image_filename)
            img_width, img_height = get_image_size(image_path)
        except KeyError as e:
            print(f"Missing key {e} in annotation: {annotation}")
            continue
        except FileNotFoundError:
            print(f"Image file not found: {image_path}")
            continue
        except Exception as e:
            print(f"Error processing annotation {annotation}: {e}")
            continue

        yolo_label_path = os.path.join(output_dir, f'{image_id}.txt')
        with open(yolo_label_path, 'w') as label_file:
            for obj in objects:
                try:
                    class_id = obj['class_id']  # Or use class_mapping
                    bbox = obj['bbox']

                    if len(bbox) == 4:
                        # Assuming [x_min, y_min, width, height]
                        x_center = (bbox[0] + bbox[2] / 2) / img_width
                        y_center = (bbox[1] + bbox[3] / 2) / img_height
                        width_norm = bbox[2] / img_width
                        height_norm = bbox[3] / img_height
                    elif len(bbox) == 4:
                        # Assuming [x_min, y_min, x_max, y_max]
                        x_min, y_min, x_max, y_max = bbox
                        width = x_max - x_min
                        height = y_max - y_min
                        x_center = (x_min + width / 2) / img_width
                        y_center = (y_min + height / 2) / img_height
                        width_norm = width / img_width
                        height_norm = height / img_height
                    else:
                        print(f"Invalid bbox format for object: {obj}")
                        continue

                    label_file.write(f'{class_id} {x_center:.6f} {y_center:.6f} {width_norm:.6f} {height_norm:.6f}\n')
                except KeyError as e:
                    print(f"Missing key {e} in object: {obj}")
                except Exception as e:
                    print(f"Error processing object {obj}: {e}")
```
